```python
# 분포 볼 때 사용할 함수
def plot_dist(x, bins, hist = True, kde = None, rug = None, label = None, color=None, line_median = None, line_mean = None, x_format = False, y_format = False):
  data = np.asarray(x)
  sns.set(rc = {"figure.figsize": (20, 10)})
  ax = sns.distplot(data, bins = int(round((abs(data.max()) - abs(data.min())) * bins,1)), hist = hist, kde = kde, rug=rug, label = label, color=color)
  
  if line_median is None and line_mean is None:
    pass
  elif line_median is not None and line_mean is None:
    plt.axvline(round(np.median(data), 2), 1, 0, color = 'red', ls = '--', alpha = 0.5)
    plt.text(x = np.median(data), y = line_median, s = "Median:" + str(round(np.median(data), 3)), bbox = dict(facecolor = 'red', alpha = 0.3), fontsize = 14)
  elif line_median is None and line_mean is not None:
    plt.axvline(round(np.mean(data), 2), 1, 0, color = 'blue', ls = '--', alpha = 0.5)
    plt.text(x = np.mean(data), y = line_mean, s = "Mean:" + str(round(np.mean(data), 3)), bbox = dict(facecolor = 'blue', alpha = 0.3), fontsize = 14)
  else:
    plt.axvline(round(np.median(data), 2), 1, 0, color = 'red', ls = '--', alpha = 0.5)
    plt.axvline(round(np.mean(data), 2), 1, 0, color = 'blue', ls = '--', alpha = 0.5)
    plt.text(x = np.median(data), y = line_median, s = "Median:" + str(round(np.median(data), 3)), bbox = dict(facecolor = 'red', alpha = 0.3), fontsize = 14)
    plt.text(x = np.mean(data), y = line_mean, s = "Mean:" + str(round(np.mean(data), 3)), bbox = dict(facecolor = 'blue', alpha = 0.3), fontsize = 14)
  if y_format is True:
    ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
  if x_format is True:
    ax.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    
def plot_title(title, xlabel, ylabel, fontproperties = None, fontsize = None):
  plt.title(label = title, fontproperties = fontproperties, fontsize = fontsize)
  plt.xlabel(xlabel = xlabel, fontproperties = fontproperties, fontsize = fontsize)
  plt.ylabel(ylabel = ylabel, fontproperties = fontproperties, fontsize = fontsize)
```

# 주성분 분석(PCA)
- Principle Components Analysis

- PCA: 수치형 변수가 어떤 식으로 공변하는지 알아내는 기법
  - 다수의 수치형 예측변수들을 더 적은 수의 변수들의 집합으로 나타내는 것
  - 전체 변수들의 <span style="color:orange">변동성</span>을 거의 대부분 설명할 수 있는 수의 변수들의 집합을 주성분이라고 함

- 주성분 분석 개념 이해를 위한 <span style="color:orange">고유 값, 고유 벡터</span> 설명


- $ A = \begin{bmatrix}
    2 & 1 \\
    1 & 2
    \end{bmatrix} $, 
    $ X = \begin{bmatrix}
    1 \\
    1
    \end{bmatrix} $
- $X$ 벡터를 좌표평면에 그리면..⬇︎

<img src="../image/ch7_eigenvalue.png" align="left">

- 벡터 $X$에 $A$라는 행렬로 <span style="color:orange">선형변환(Linear transformation)</span>을 가정하면..  
  - $ AX =
    \begin{bmatrix}
    2 & 1 \\
    1 & 2
    \end{bmatrix}
    \begin{bmatrix}
    1 \\
    1
    \end{bmatrix} =
    \begin{bmatrix}
    2+1 \\
    1+2
    \end{bmatrix} =
    \begin{bmatrix}
    3 \\
    3
    \end{bmatrix} $  
  - 선형변환은 사실상 행렬의 곱과 같은데, 단순히 곱하는 방법을 알기 보다는 기하학적으로 어떻게 변화하는 것인지 아는 게 중요함.

<img src="../image/ch7_eigenvalue_2.png" align="left">

- 이게 무슨뜻일까? 어느 벡터$X$에 행렬$A$를 곱하니 <span style="color:orange">벡터의 방향은 그대로이고 길이만 변하였음</span>
- 이를 다르게 표현하면
  - $ AX = \lambda X $
  - $ (A-\lambda I)X=0 $ ($I$=Identity Matrix) ⇢ 이 식이 성립하려면 $ det(A-\lambda I)=0 $
  - $ det \begin{bmatrix}
          2-\lambda & 1 \\
          1 & 2-\lambda
          \end{bmatrix} = 0 $
  - $ {(2-\lambda)^2} - 1 = (\lambda-3)(\lambda-1) $
  - $ \lambda_1 = 3, \lambda_2 = 1 $
  
- 고유값은 3과 1이므로 $AX = \lambda X$를 이용해 고유벡터를 구하면..
  - $ \lambda_1 = 3 $
    - $ \begin{bmatrix}
        2 & 1 \\
        1 & 2
        \end{bmatrix}
        \begin{bmatrix}
        X_1 \\
        X_2
        \end{bmatrix} =
        3
        \begin{bmatrix}
        X_1 \\
        X_2
        \end{bmatrix} $  
    - $ 2X_1 + X_2 = 3X_1 $
    - $ X_1 + 2X_2 = 3X_2 $
    - $ X_1 = 1, X_2 = 1 $
    - $ X_1 = X_2 $
    - 고유값이 3일 때 고유벡터는 $ \begin{bmatrix} 1 \\ 1 \end{bmatrix} $ 라는 것임
  - $ \lambda_1 = 1 $
    - $ \begin{bmatrix}
        2 & 1 \\
        1 & 2
        \end{bmatrix}
        \begin{bmatrix}
        X_1 \\
        X_2
        \end{bmatrix} =
        1
        \begin{bmatrix}
        X_1 \\
        X_2
        \end{bmatrix} $  
    - $ 2X_1 + X_2 = X_1 $
    - $ X_1 + 2X_2 = X_1 $
    - $ X_1 = 1, X_2 = -1 $
    - $ X_1 = -X_2 $
    - 고유값이 1일 때 고유벡터는 $ \begin{bmatrix} 1 \\ -1 \end{bmatrix} $ 라는 것임
- 결국 우리가 가진 데이터를 공분산 행렬로 바꾸게 된다면..
  - 예를 들어 공분산 행렬이 $ \sum = \begin{bmatrix}
                        2 & 1 \\
                        1 & 2
                        \end{bmatrix} $ 일 때  
  - 공분산 행렬에 선형변환을 했을 때 그 크기는 변하고 <span style="color:red">방향이 변하지 않는</span> 벡터가 존재한다면, 그 벡터의 크기는 각 1배와 3배가 된다는 의미임!
  - 만약 방향이 바뀌게 된다면 공분산 행렬의 모양 자체가 바뀐다는 것이므로, 원래의 분산을 유지한채 크기만 크기만 변화시키는 벡터, 즉 <span style="color:red">사영(Projection)할 수 있는 새로운 축</span>을 찾아내겠다는 것임. 이로써 두 차원이 하나의 차원으로 감소하게 되는 것임.
  - 이 때, 각 변수의 분산은 2와 2로 합이 4로 총 분산은 4가 된다. 그런데 고유값이 1과 3인데, 이것 역시 두 변수의 분산을 담고 있는 정보가 되는 것임!
  - <span style="color:orange">고유값의 합(1+3=4)은 두 분산의 합과 같음(2+2=4)</span>
  - 또한, 고유값의 곱(1x3=3)은 두 $det\sum$ = (2x2-1x1=3)으로 동일
- 그래서 이게 의미하는 바가 뭘끼?
  - 고유값의 합은 각 변수의 분산의 합과 같다는 점에서, <span style="color:orange">고유 값이 클수록 설명력이 높은 주성분(PC)</span>이라는 것을 알 수 있음
  - 또한, <span style="color:orange">주성분의 분산이 각 변수의 원래 분산 보다도 작다면 오히려 설명력이 더 낮은 것이므로 쓸모 없는 PC가 되는 것임</span>
  - 이를 통해 <span style="color:orange">우리가 몰랐던 PC를 알게 된다는 것을 알 수 있게 되는 것</span>이며, 최대한 설명력이 높은 주성분만 사용함으로써 차원을 축소하는 효과를 가져오게 되는 것임
- $ \sum X = \lambda X $

            \end{bmatrix} $
- $ \sum X = \lambda X $
- det($ \begin{bmatrix}
        1-\lambda & 0.98 \\
        0.98 & 1-\lambda
        \end{bmatrix} = 0$ ⇢ $ (1-\lambda)^2 - 0.98^2 $
- $ \lambda_1 = 1.98, \lambda_2 = 0.02 $
- $ \begin{bmatrix}
    1 & 0.98 \\
    0.98 & 1
    \end{bmatrix}
    \begin{bmatrix}
    X_1 \\
    X_2
    \end{bmatrix} = 
    \lambda \begin{bmatrix}
    X_1 \\
    X_2
    \end{bmatrix} $에 의해 고유벡터를 구하면..
  - $ \lambda = 1.98 $
    - $ X_1 + 0.98 X_2 = 1.98 X_1 $
    - $ 0.98 X_1 + X_2 = 1.98 X_2 $
    - $ e_1 = \begin{bmatrix}
              1 \\
              1
              \end{bmatrix} $
  - $ \lambda = 0.02 $
    - $ X_1 + 0.98 X_2 = 0.02 X_1 $
    - $ 0.98 X_1 + X_2 = 0.02 X_2 $
    - $ e_1 = \begin{bmatrix}
              1 \\
              -1
              \end{bmatrix} $
  - 고유값이 1.98일 때, 두 변수의 분산을 1.98/2 = 99% 만큼 설명함
  - 고유값이 0.02일 때, 두 변수의 분산을 0.02/2 = 1% 만큼 설명함

- 그렇다면 PC1, PC2라는 새로분 변수는 어떻게 만들어지는가?
- 앞에서 만들어 놓았던 표준화된 각 변수에, 고유 벡터를 곱하여 더해주면 된다.
  - 즉, 고유 벡터의 각 스칼라 값이 <span style="color:orange">Weight</span>가 되는 셈임!
- 위에서 고유값$(\lambda)$이 3일 때 고유벡터는 $ e_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $ 이었으므로,
  - PC1 =  국어x1 + 영어x1
- 또, 고유값$(\lambda)이$ 1일 때 고유벡터는 $ e_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix} $ 이었으므로,
  - PC2 = 국어x1 + 영어x(-1)
- 이렇게 구한 각 PC1, PC2의 분산을 구해보면 분산의 합이 4가 된다는 것을 확인할 수 있다.

```python
df_z['PC1'] = (df_z['국어'] * 1 ) + (df_z['영어'] * 1 )
df_z['PC2'] = (df_z['국어'] * 1 ) + (df_z['영어'] * -1 )
```

- 마지막으로.. 주성분이라는 것이 상관관계가 높은 변수로 인해 만들어지는 값이라고 했으니 원래의 변수와 PC와의 상관관계가 어느 정도일지 확인이 필요하다.
- 아래의 결과를 보면 PC1과 국어, 그리고 PC1과 영어는 아주 높은 상관관계를 가지며,
- 반면 PC2와 국어, 그리고 PC2와 영어는 낮은 상관관계를 가진다는 것 역시 확인할 수 있다.
- 물론 상관관계라는 것이 상대적이긴 하나, PC1과 비교해보면 PC2의 상관관계는 매우 낮다는 것이다.

```python
df_z.corr()
```

- 그렇다면 PCA는 항상 해야하는 것인가? 어떻게 판단해야 할까?
- 주성분분석에 적합한 데이터인지를 먼저 확인해야 하는데, 이를 위해 구형성 검증(bartlett's test of sphericity)을 해야한다.
  - 데이터를 scatter로 뿌려보았을 때 퍼진 정도가 원형으로 생겼다면 상관관계가 0이기 때문에 가능한 PC를 찾기가 어렵다는 것임.
- 구형성 검증 예시
  - $H_0$: 상관관계가 0이다 (상관관계 행렬 = Identity 행렬)
  - $H_0$: 상관관계가 0이 아니다 (상관관계 행렬 != Identity 행렬)
  - $ \begin{bmatrix}
      r_{11} & r_{12} & r_{13} \\
      r_{21} & r_{22} & r_{23} \\
      r_{31} & r_{32} & r_{33}
      \end{bmatrix} = 
      \begin{bmatrix}
      1 & 0 & 0 \\
      0 & 1 & 0 \\
      0 & 0 & 1
      \end{bmatrix} $ ⇢ 만약 이렇다면 주성분은 존재하지 않음

```python
def abline(slope, intercept, ax):
    """Calculate coordinates of a line based on slope and intercept"""
    x_vals = np.array(ax.get_xlim())
    return (x_vals, intercept + slope * x_vals)

ax = oil_px.plot.scatter(x='XOM', y='CVX', alpha=0.3, figsize=(8, 8))
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.plot(*abline(loadings.loc[0, 'CVX'] / loadings.loc[0, 'XOM'], 0, ax),
        '--', color='C1')
ax.plot(*abline(loadings.loc[1, 'CVX'] / loadings.loc[1, 'XOM'], 0, ax),
        '--', color='C1')

plt.tight_layout()
plt.show()
```

## 주성분 계산

- 표준화를 해주고 계산해보기

```python
oil_px_z = oil_px.copy()

for col in oil_px_z.columns:
  oil_px_z[col] = oil_px_z[col].apply(lambda x: (x-oil_px_z[col].mean())/oil_px_z[col].std())
  
oil_px_z
```

```python
oil_px_z.var()
```

```python
np.cov(oil_px_z.T.to_numpy())
```

- $ (1- \lambda)^2 - 0.79^2 = 0 $
- $ (1- \lambda -0.79)(1- \lambda + 0.79) = 0$
- $ (0.21 - \lambda)(1.79 - \lambda) = 0$

## K-Means 알고리즘

```python
syms = sorted(['AAPL', 'MSFT', 'CSCO', 'INTC', 'CVX', 'XOM', 'SLB', 'COP', 
               'JPM', 'WFC', 'USB', 'AXP', 'WMT', 'TGT', 'HD', 'COST'])
top_sp = sp500_px.loc[sp500_px.index >= '2011-01-01', syms]
kmeans = KMeans(n_clusters=5).fit(top_sp)
```

## Interpreting the Clusters

```python
from collections import Counter
print(Counter(kmeans.labels_))
```

```python
centers = pd.DataFrame(kmeans.cluster_centers_, columns=syms)

f, axes = plt.subplots(5, 1, figsize=(8, 8), sharex=True)
for i, ax in enumerate(axes):
    center = centers.loc[i, :]
    maxPC = 1.01 * np.max(np.max(np.abs(center)))
    colors = ['C0' if l > 0 else 'C1' for l in center]
    ax.axhline(color='#888888')
    center.plot.bar(ax=ax, color=colors)
    ax.set_ylabel(f'Cluster {i + 1}')
    ax.set_ylim(-maxPC, maxPC)

plt.tight_layout()
plt.show()
```

## Selecting the Number of Clusters

```python
inertia = []
for n_clusters in range(2, 14):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(top_sp)
    inertia.append(kmeans.inertia_ / n_clusters)
inertias = pd.DataFrame({'n_clusters': range(2, 14), 'inertia': inertia})
ax = inertias.plot(x='n_clusters', y='inertia')
plt.xlabel('Number of clusters(k)')
plt.ylabel('Average Within-Cluster Squared Distances')
plt.ylim((0, 1.1 * inertias.inertia.max()))
ax.legend().set_visible(False)

plt.tight_layout()
plt.show()
```

## 모델 기반 클러스터링

```python
mean = [0.5, -0.5]
cov = [[1, 1], [1, 2]]
probability = [.5, .75, .95, .99]
def probLevel(p):
    D = 1
    return (1 - p) / (2 * math.pi * D)
levels = [probLevel(p) for p in probability]

fig, ax = plt.subplots(figsize=(5, 5))

x, y = np.mgrid[-2.8:3.8:.01, -5:4:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y
rv = multivariate_normal(mean, cov)


CS = ax.contourf(x, y, rv.pdf(pos), cmap=cm.GnBu, levels=50)
ax.contour(CS, levels=levels, colors=['black'])
ax.plot(*mean, color='black', marker='o')


plt.tight_layout()
plt.show()
```

## 클러스터 개수 결정하기

```python
results = []
covariance_types = ['full', 'tied', 'diag', 'spherical']
for n_components in range(1, 9):
    for covariance_type in covariance_types:
        mclust = GaussianMixture(n_components = n_components, warm_start=True,
                                 covariance_type = covariance_type)
        mclust.fit(df)
        results.append({
            'bic': mclust.bic(df),
            'n_components': n_components,
            'covariance_type': covariance_type,
        })
        
results = pd.DataFrame(results)

colors = ['C0', 'C1', 'C2', 'C3']
styles = ['C0-','C1:','C0-.', 'C1--']

fig, ax = plt.subplots(figsize=(8, 8))
for i, covariance_type in enumerate(covariance_types):
    subset = results.loc[results.covariance_type == covariance_type, :]
    subset.plot(x='n_components', y='bic', ax=ax, label=covariance_type, 
                kind='line', style=styles[i]) # , color=colors[i])

plt.tight_layout()
plt.show()
```
