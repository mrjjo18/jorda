```python
# 분포 볼 때 사용할 함수
def plot_dist(x, bins, hist = True, kde = None, rug = None, label = None, color=None, line_median = None, line_mean = None, x_format = False, y_format = False):
  data = np.asarray(x)
  sns.set(rc = {"figure.figsize": (20, 10)})
  ax = sns.distplot(data, bins = int(round((abs(data.max()) - abs(data.min())) * bins,1)), hist = hist, kde = kde, rug=rug, label = label, color=color)
  
  if line_median is None and line_mean is None:
    pass
  elif line_median is not None and line_mean is None:
    plt.axvline(round(np.median(data), 2), 1, 0, color = 'red', ls = '--', alpha = 0.5)
    plt.text(x = np.median(data), y = line_median, s = "Median:" + str(round(np.median(data), 3)), bbox = dict(facecolor = 'red', alpha = 0.3), fontsize = 14)
  elif line_median is None and line_mean is not None:
    plt.axvline(round(np.mean(data), 2), 1, 0, color = 'blue', ls = '--', alpha = 0.5)
    plt.text(x = np.mean(data), y = line_mean, s = "Mean:" + str(round(np.mean(data), 3)), bbox = dict(facecolor = 'blue', alpha = 0.3), fontsize = 14)
  else:
    plt.axvline(round(np.median(data), 2), 1, 0, color = 'red', ls = '--', alpha = 0.5)
    plt.axvline(round(np.mean(data), 2), 1, 0, color = 'blue', ls = '--', alpha = 0.5)
    plt.text(x = np.median(data), y = line_median, s = "Median:" + str(round(np.median(data), 3)), bbox = dict(facecolor = 'red', alpha = 0.3), fontsize = 14)
    plt.text(x = np.mean(data), y = line_mean, s = "Mean:" + str(round(np.mean(data), 3)), bbox = dict(facecolor = 'blue', alpha = 0.3), fontsize = 14)
  if y_format is True:
    ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
  if x_format is True:
    ax.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    
def plot_title(title, xlabel, ylabel, fontproperties = None, fontsize = None):
  plt.title(label = title, fontproperties = fontproperties, fontsize = fontsize)
  plt.xlabel(xlabel = xlabel, fontproperties = fontproperties, fontsize = fontsize)
  plt.ylabel(ylabel = ylabel, fontproperties = fontproperties, fontsize = fontsize)
```

# 나이브 베이즈

## Non-naive

- 베이지안 분류 이해를 위해 <span style="color:orange">'나이브하지 않은'</span> 분류를 생각해보자.
  - 1) 예측변수 값이 동일한 모든 row를 찾는다.
  - 2) 해당 row들이 가장 많이 속한 클래스를 찾는다.
  - 3) 위의 row와 동일한 새로운 row가 들어왔을 때 해당 클래스로 분류한다.  
  
- 나이브하지 않은 베이지언 분류는 왜 현실성이 없을까?

## Naive

- '나이브한' 분류
  - 1) 분류하고자 하는 클래스를 각 예측변수로 조건부확률 $P(X_i|Y$=$i)$을 구한다.
  - 2) 각 확률값을 곱한 다음 $Y$=$i$에 속한 row의 비율을 곱한다. 
  - 3) 모든 클래스에 대해 1) ~ 2) 단계를 반복한다.
  - 4) 2)에서 모든 클래스에 대해 구한 확률 값을 모두 더한 값으로 클래스 $i$의 확률을 나누면 결과 $i$의 확률을 구할 수 있다. (여기를 이해하는 것이 중요..)
  - 5) 이 예측변수에 대해 가장 높은 확률을 갖는 클래스를 해당 row에 할당한다.
  
- 확률 식
  - $ P(X_1, X_2, ... ,X_p) = P(Y=0)(P(X_1|Y=0)P(X_2|Y=0)...P(X_p|Y=0)) + P(Y=1)(P(X_1|Y=1)P(X_2|Y=1)...P(X_p|Y=1)) $
    - 사실 $ P(X_1, X_2, ... ,X_p)$ 값은 확률이 Y에 독립이면서 0과 1사이에 오도록 하기 위한 스케일링 계수임.
                              
  - 왜 이 공식을 '나이브'라고 부르는 걸까?
    - <span style="color:orange">$ P(X_1, X_2, ..., X_p|Y=i)가 아닌 P(X_j|Y=i)$ 를 추정하는 이유는 무엇일까?</span>
    
- 나이브 베이즈는 베이지언 통계 방법으로 간주되지 않는다. 상대적으로 통계 지식이 거의 필요 없는 데이터 중심의 경험적 방법이다. 베이즈 규칙과 비슷한 예측 계산이 들어가다 보니 이름을 그렇게 붙였을 뿐이다.

```python
loan_data = pd.read_csv('./data/loan_data.csv.gz', compression='gzip', index_col=0)
```

```python
df = loan_data[['purpose_', 'home_', 'emp_len_', 'outcome']].copy()
df.head()
```

```python
print("purpose_ :", df.purpose_.unique(), '\n',
      "home_ :", df.home_.unique(), '\n',
      "emp_len_ :", df.emp_len_.unique())
```

```python
print('P(X=RENT|Y=default)', df[(df.outcome=='default')&(df.home_=='RENT')].shape[0] / df[df.outcome=='default'].shape[0])
print('P(X=OWN|Y=default)', df[(df.outcome=='default')&(df.home_=='OWN')].shape[0] / df[df.outcome=='default'].shape[0])
print('P(X=MORTGAGE|Y=default)', df[(df.outcome=='default')&(df.home_=='MORTGAGE')].shape[0] / df[df.outcome=='default'].shape[0])
```

```python
result = pd.DataFrame()
for y in df.outcome.unique():
  for col in df.columns[df.columns!='outcome']:
    for x in df[col].unique():
      idx = 'P({0} | {1})'.format(x, y)
      p = df[(df[col]==x)&(df.outcome==y)].shape[0] / df[df.outcome==y].shape[0]
      df_tmp = pd.DataFrame({'P':p}, index=[idx])
      result = result.append(df_tmp)
result = result.reset_index()
result.columns=['condition', 'prob']
result
```

```python
# convert to categorical
loan_data.outcome = loan_data.outcome.astype('category')
loan_data.outcome.cat.reorder_categories(['paid off', 'default'])
loan_data.purpose_ = loan_data.purpose_.astype('category')
loan_data.home_ = loan_data.home_.astype('category')
loan_data.emp_len_ = loan_data.emp_len_.astype('category')

predictors = ['purpose_', 'home_', 'emp_len_']
outcome = 'outcome'
X = pd.get_dummies(loan_data[predictors], prefix='', prefix_sep='')
y = loan_data[outcome]

naive_model = MultinomialNB(alpha=0.01, fit_prior=True)
naive_model.fit(X, y)

new_loan = X.loc[146:146, :] 
print('predicted class: ', naive_model.predict(new_loan)[0])

probabilities = pd.DataFrame(naive_model.predict_proba(new_loan),
                             columns=naive_model.classes_)
print('predicted probabilities',)
print(probabilities)
```

## 공분산행렬

- $ {\displaystyle s_{x,z} = \frac{\sum_{i=1}^n (x_i-\bar{x})(z_i-\bar{z})}{n-1}} $

- $ Cov = \begin{bmatrix}
    {s_x}^2 & s_{x, z} \\
    s_{x, z} & {s_z}^2
    \end{bmatrix} $
    
- 공분산($S_{x,z}$)의 <span style="color:orange">의미</span>는 무엇일까?
  - 변수 1개의 분산 Vs. 변수 2개의 분산(?)

  - $ {\displaystyle corr = \frac{Cov(x,y)}{\sqrt{Var(x) * Var(y)}} } $

## 피셔의 선형판별

- 집단 간 <span style="color:orange">분산</span>은 최대화하고, 집단 내 <span style="color:orange">분산</span>은 최소화 하는 것이 핵심임.
- 즉, 데이터를 특정한 축에 사영(projection)한 후 두 범주를 잘 구분할 수 있는 (직)선을 찾는 걸 목표로 함.
  - ${\displaystyle \frac{SS_{사이}}{SS_{내부}} = \frac{(m_1-m_2)^2}{s_1^2 + s_2^2} }$  
  
[https://ratsgo.github.io/machine%20learning/2017/03/21/LDA/]  
[elearning.kocw.net/document/lec/2012/DukSung/KimJaehee/13.pdf]

## 예시

```python
loan3000 = pd.read_csv('./data/loan3000.csv', index_col=0).reset_index(drop=True)
```

```python
loan3000.outcome = loan3000.outcome.astype('category')

predictors = ['borrower_score', 'payment_inc_ratio']
outcome = 'outcome'

X = loan3000[predictors]
y = loan3000[outcome]

loan_lda = LinearDiscriminantAnalysis()
loan_lda.fit(X, y)
print(pd.DataFrame(loan_lda.scalings_, index=X.columns))
```

```python
pred = pd.DataFrame(loan_lda.predict_proba(loan3000[predictors]),
                    columns=loan_lda.classes_)
print(pred.head())
```

```python
# Use scalings and center of means to determine decision boundary
center = np.mean(loan_lda.means_, axis=0)
slope = - loan_lda.scalings_[0] / loan_lda.scalings_[1]
intercept = center[1] - center[0] * slope

# payment_inc_ratio for borrower_score of 0 and 20
x_0 = (0 - intercept) / slope
x_20 = (20 - intercept) / slope

lda_df = pd.concat([loan3000, pred['default']], axis=1)
lda_df.head()

fig, ax = plt.subplots(figsize=(8, 8))
g = sns.scatterplot(x='borrower_score', y='payment_inc_ratio',
                    hue='default', data=lda_df,
#                     palette=sns.diverging_palette(240, 10, n=9, as_cmap=True),
                    ax=ax, legend=False)

ax.set_ylim(0, 20)
ax.set_xlim(0.15, 0.8)
ax.plot((x_0, x_20), (0, 20), linewidth=3)
ax.plot(*loan_lda.means_.transpose())

plt.tight_layout()
plt.show()
```

# 로지스틱 회귀

- 종속변수가 이항변수일 때 OLS에 기반한 기존 회귀모형은 절대로 사용하면 안 되는 걸까? 만약 그렇다면 그 이유는 무엇일까?
  - X가 1 증가할 때 Y가 b만큼 증가(?)

## 로지스틱 반응 함수와 로짓

- 종속변수가 0, 1 두 가지일 때, 1이 될 <span style="color:orange">확률</span>을 $p$라고 하자.
- $ p = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_qx_q $ ➠ $p$가 0~1 사이로 딱 떨어지는가?
- $\displaystyle{ p = \frac{1}{1+e^{-(\beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_qx_q)}} } $  
   
- $ \displaystyle{ odds = \frac{p}{1-p} } $
- $ \displaystyle{ p = \frac{odds}{1+odds} } $
- $ \displaystyle{ odds(Y=1)=e^{\beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_qx_q} } $
- $ \displaystyle{ log(odds(Y=1))=\beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_qx_q } $

<img src="../image/sigmoid.png" align="left">

<오즈비 예시>

Push | 주문 O | 주문 X | 합계
---|---|---|---
A | 1명 | 1,999명 | 2,000명
B | 1명 | 7,999명 | 8,000명  

- Push A 고객의 주문 오즈(odds) = $ \displaystyle{ \frac{ \frac{1}{2000} } { 1-\frac{1}{2000} } = \frac{1}{1999} } $
- Push B 고객의 주문 오즈(odds) = $ \displaystyle{ \frac{ \frac{1}{8000} } { 1-\frac{1}{8000} } = \frac{1}{7999} } $
- Push B 대비 Push A 주문 오즈비 (odds ratio) = $ \displaystyle{ \frac{ \frac{1}{1999} } { \frac{1}{7999} } = 4.0015 } $
- Push A의 주문 확률(1/2000)과 Push B의 주문 확률(1/8000)은 거의 차이가 없음.
- Push B로 1명이 주문하면 Push A로는 4명 정도가 주문한다.
  - 오즈를 오즈로 나누면 비교가 가능해진다.
---
- p ≒ 0 ⇢ odds ratio ≒ 0
- p ≒ 1 ⇢ odds ratio ≒ ∞
- $ \displaystyle{ 0 < \frac{p}{1-p} < ∞ } $
- <span style="color:orange"> $ \displaystyle{  -∞ < log_e(\frac{p}{1-p}) < +∞ } $ </span> ➠ 이것을 통해 얻을 수 있는 이점이 무엇인지 아는 게 중요
- 로짓 = $ \displaystyle{ log_e(\frac{p}{1-p}) }$

```python
p = np.arange(0.01, 1, 0.01)
df = pd.DataFrame({
    'p': p,
    'logit': np.log(p / (1 - p)),
    'odds': p / (1 - p),
})

fig, ax = plt.subplots(figsize=(8, 8))
ax.axhline(0, color='grey', linestyle='--')
ax.axvline(0.5, color='grey', linestyle='--')
ax.plot(df['p'], df['logit'])
ax.set_xlabel('Probability')
ax.set_ylabel('logit(p)')

plt.tight_layout()
plot_title(title="", xlabel="Probability", ylabel="logit(p)", fontproperties=fontprop, fontsize=13)
plt.show()
```

## 로지스틱 회귀와 GLM

- 앞서 유도한 로지스틱 회귀방정식에서 응답변수는 1의 이진 출력에 대한 로그 오즈 값이다.

```python
from statsmodels.genmod.generalized_linear_model import GLMResults
def partialResidualPlot(model, df, outcome, feature, fig, ax):
    y_actual = [0 if s == 'default' else 1 for s in df[outcome]]
    y_pred = model.predict(df)
    org_params = model.params.copy()
    zero_params = model.params.copy()
    # set model parametes of other features to 0
    for i, name in enumerate(zero_params.index):
        if feature in name:
            continue
        zero_params[i] = 0.0
    model.initialize(model.model, zero_params)
    feature_prediction = model.predict(df)
    ypartial = -np.log(1/feature_prediction - 1)
    ypartial = ypartial - np.mean(ypartial)
    model.initialize(model.model, org_params)
    results = pd.DataFrame({
        'feature': df[feature],
        'residual': -2 * (y_actual - y_pred),
        'ypartial': ypartial/ 2,
    })
    results = results.sort_values(by=['feature'])

    ax.scatter(results.feature, results.residual, marker=".", s=72./fig.dpi)
    ax.plot(results.feature, results.ypartial, color='black')
    ax.set_xlabel(feature)
    ax.set_ylabel(f'Residual + {feature} contribution')
    return ax

fig, ax = plt.subplots(figsize=(8, 8))
partialResidualPlot(results, loan_data, 'outcome', 'payment_inc_ratio', fig, ax)
ax.set_xlim(0, 25)
ax.set_ylim(-2.5, 2.5)


plt.tight_layout()
plt.show()
```

## Precision, Recall, and Specificity

<img src="../image/confusion_matrix.png" align="left" width="800">

- precision = $ \displaystyle{ \frac{TP}{TP + FP} } $  
   - 
- recall = $ \displaystyle{ \frac{TP}{TP + FN} } $  
   - 
- Specificity = $ \displaystyle{ \frac{TN}{FP + TN} } $  
  - 
- accuracy = $ \displaystyle{ \frac{TP + TN}{ALL} } $  

```python
conf_mat = confusion_matrix(y, logit_reg.predict(X))
print('Precision', conf_mat[0, 0] / sum(conf_mat[:, 0]))
print('Recall', conf_mat[0, 0] / sum(conf_mat[0, :]))
print('Specificity', conf_mat[1, 1] / sum(conf_mat[1, :]))
```

```python
default_wt = 1 / np.mean(full_train_set.outcome == 'default')
wt = [default_wt if outcome == 'default' else 1 for outcome in full_train_set.outcome]

full_model = LogisticRegression(penalty="l2", C=1e42, solver='liblinear')
full_model.fit(X, y, wt)
print('percentage of loans predicted to default (weighting): ', 
print(      100 * np.mean(full_model.predict(X) == 'default')))
```

```python
models = {
    'Decision Tree': loan_tree,
    'Linear Discriminant Analysis': loan_lda,
    'Logistic Regression': logit_reg,
    'Generalized Additive Model': gam,
}

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

xvalues = np.arange(0.25, 0.73, 0.005)
yvalues = np.arange(-0.1, 20.1, 0.1)
xx, yy = np.meshgrid(xvalues, yvalues)
X = np.c_[xx.ravel(), yy.ravel()]

boundary = {}

for n, (title, model) in enumerate(models.items()):
    ax = axes[n // 2, n % 2]
    predict = model.predict(X)
    if 'Generalized' in title:
        Z = np.array([1 if z > 0.5 else 0 for z in predict])
    else:
        
        Z = np.array([1 if z == 'default' else 0 for z in predict])
    Z = Z.reshape(xx.shape)
    boundary[title] = yvalues[np.argmax(Z > 0, axis=0)]
    boundary[title][Z[-1,:] == 0] = yvalues[-1]

    c = ax.pcolormesh(xx, yy, Z, cmap='Blues', vmin=0.1, vmax=1.3)
    ax.set_title(title)
    ax.grid(True)

plt.tight_layout()
plt.show()
```
